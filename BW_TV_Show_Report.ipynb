```python
# Example DataFrame
data = {
    'Season': [1, 2, 3, 4, 5],
    'Viewership': [1.2, 1.5, 1.1, 1.6, 1.4]  # millions
}
df = pd.DataFrame(data)

# Calculate change in viewership
df['Change'] = df['Viewership'].diff()

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(df['Season'], df['Viewership'], marker='o', linestyle='-')
plt.title('Season-to-Season Viewership Changes')
plt.xlabel('Season')
plt.ylabel('Viewership (millions)')
plt.grid(True)
plt.xticks(df['Season'])  # Ensure all seasons are labeled
plt.show()
```

The viewership peaked in Season __4__ at __1.6 million__, an increase from __1.1 million__ in Season __3__. However, there was a decrease in viewership by __0.2 million__ from Season __4__ to Season __5__.
